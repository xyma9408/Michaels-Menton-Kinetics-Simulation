```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Define Lineweaverâ€“Burk plot model  
Visualize the plot where Km = 10 and vmax = 1

```python
def model(S_reverse, Km, vmax):
    return Km/vmax * S_reverse + 1/vmax

S = np.arange(1, 200, 1)
S_reverse = 1/S
v_reverse = model(S_reverse, 10, 1)
plt.plot(S_reverse, v_reverse)
```

```python
S_exp = np.array([5, 10, 15, 20, 25, 30 ,35, 40, 45, 50, 60, 70, 80, 90, 100, 125, 150, 200])
S_exp_reverse = 1/S_exp
v_calc_reverse = model(S_reverse = S_exp_reverse, Km = 10, vmax = 1)
v_calc = 1/v_calc_reverse

def create_data(n, std):
    result = np.zeros((n, len(S_exp)))
    for index, e in enumerate(v_calc):
        temp = std * np.random.randn(n) + e
        result[:, index] = temp
    return np.array(result)

sim_data = create_data(100, 0.03)
sim_data_reverse = 1/sim_data

for yi in sim_data_reverse:
    plt.scatter(S_exp_reverse, yi)
```
